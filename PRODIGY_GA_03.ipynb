```python
sample_text = """
Artificial intelligence is transforming the world.
Machine learning allows computers to learn from data.
Fine-tuning models like GPT-2 helps them understand specific styles.
This is a test dataset for Task-01.
""" * 100

with open("your_custom_data.txt", "w") as f:
    f.write(sample_text)

print("File 'your_custom_data.txt' created successfully!")
```

```python
import markovify

with open("your_custom_data.txt") as f:
    text = f.read()

text_model = markovify.Text(text, state_size=1)

print("--- Generated Sentences ---")
for i in range(5):
    sentence = text_model.make_sentence(test_output=False)
    if sentence:
        print(f"{i+1}: {sentence}")
    else:
        print(f"{i+1}: {text_model.make_short_sentence(max_chars=100, test_output=False)}")
```
